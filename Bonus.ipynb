```python
import pandas as pd
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
f = open('capacitor_dataset.pickle' ,'rb')
train_datasets = pickle.load(f)
print(train_datasets)
f.close()
```

```python
log_dataset = [[data[0] , np.log(data[1])] for data in train_datasets]
print(log_dataset)
```

```python
x_train = []
y_train = []
train_dataset_np = np.array(log_dataset)
print()
x_train = [[data[0]] for data in train_dataset_np]
y_train = [[data[1]] for data in train_dataset_np]
# print(train_dataset_np)
print(x_train)
print(y_train)
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
weight = model.coef_
bias = model.intercept_
print(weight , bias)
```
